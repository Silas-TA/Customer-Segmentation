```python
# imports
import pandas as pd
import numpy as np
```

```python
# Initialize the data
df = pd.read_csv(r"online_retail_II.csv")
df = df.drop_duplicates() #delete duplicates
df.head(5)
```

```python
# dropping description column an dempty columns
df.drop(columns= "Description", inplace=True)
df.dropna(subset=['Customer ID'], inplace = True)
df.head(5)
```

```python
# remove voided invoices
new_df = df[~df["Invoice"].str.startswith('C', na=False)]
void_df = df[df["Invoice"].str.startswith('C', na=False)]
new_df.head(10)
# all_df.head(10)
```

```python
# remove quantity and prices with negative values
new_df = new_df[(new_df["Quantity"] > 0) & (new_df["Price"] > 0)]
void_df = void_df[(void_df["Quantity"] > 0) & (void_df["Price"] > 0)]
new_df.head(10)
```

```python
save_newdf = 'cleaned_online_retail.csv'# Replace with the desired output path
save_voiddf = 'cleaned_online_retail_void.csv' 
new_df.to_csv(save_newdf, index=False)
void_df.to_csv(save_voiddf, index=False)

```
